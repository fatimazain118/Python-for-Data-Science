SeaBorn is lib in python used for visualisation and is build on matplotlib

```python
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
Fmri = sns.load_dataset("fmri")  #fmri is a inbuild dataset in seaborn
Fmri.head()
```

```python
Fmri.shape
```

Line PLot

```python
sns.lineplot(x="timepoint", y ="signal" ,data = Fmri)   #lineplot is a method in seaborn
plt.show()
```

Grouping Data with "Hue"

```python
sns.lineplot(x="timepoint", y= "signal", data = Fmri,hue = "event")    #hue determine the color of graph here we have taking colomn event's two category
plt.show()
```

```python
sns.lineplot(x="timepoint", y= "signal", data = Fmri, hue = "event", style = "event")
plt.show()
```

```python
sns.lineplot(x="timepoint", y= "signal", data = Fmri, hue = "event", style = "event", markers = True)
plt.show()
```

Histogram/Distplot while using SeaBorn we directly don't get histogram else we get distribution plot in which we have histogram and frequency curve together we call this distplot

```python
diamond = pd.read_csv("/content/gdrive/MyDrive/diamonds.csv")
diamond
```

```python
diamond.head()
```

```python
sns.distplot(diamond["price"],color = "green")   #when we want to see numerical column distribution we use histogram or frequency polygon
plt.show()
```

```python
sns.distplot(diamond["price"], hist= False)
plt.show()
```

```python
sns.distplot(diamond["price"], color ="red", bins =10, kde= False)    #kde is use to remove frequency curve and to increase the width we use bins attribute
plt.show()
```

```python
sns.distplot(diamond["price"],color ="orange", vertical = True)
plt.show() #to get plot along x-axis
```

JointPlot(Scatter plot + histogram)

```python
sns.jointplot(x="SepalLengthCm", y= "PetalLengthCm",color="olive", data = iris)
plt.show()
```

```python
sns.jointplot(x="SepalLengthCm", y= "PetalLengthCm",color="brown", data = iris, kind= "reg")  #to add regression line though scatter plot and histogram we use kind attribute
plt.show()
```

BoxPlot-- use in charactor(x-axis) and numerical variable

```python
churn = pd.read_csv("/content/gdrive/MyDrive/churn.csv")
churn.head()
```

```python
churn.columns
```

```python
sns.boxplot(x="Churn", y="tenure",data= churn)   #yes(churn = 1) means customer will leave you
plt.show()          #for comparison we use median values
```

```python
sns.boxplot(x="InternetService", y= "MonthlyCharges", data = churn, palette = "Set1")
plt.show()
```

```python
sns.boxplot(x="InternetService", y= "tenure", data = churn, color = "brown")
plt.show()
```

```python
sns.boxplot(x="Contract", y= "tenure", data = churn, linewidth = 3)  #linewidth gives thickness of boundaries
plt.show()
```

```python
sns.boxplot(x="Contract", y= "tenure", data = churn, linewidth = 3, order =["Two year","Month-to-month","One year"]) #to change the default order
plt.show()
```

```python
sns.boxplot(x="Contract", y= "tenure", data = churn, linewidth = 3, hue = "PaymentMethod") 
plt.show()
```
